```
# Start writing code here...import pandas as pd
import requests
import numpy as np
import pandas as pd
from bs4 import BeautifulSoup
```

```
url = "https://www.imdb.com/search/title/?year=2016&title_type=feature&"
response = requests.get(url)
soup = BeautifulSoup(response.content, 'html.parser')
```

```
print(soup)
```

```
movie_name = []
year = []
rating = []
genre = []
metascore = []
runtime = []
votes = []
gross = []
actors = []
```

```
movie_data = soup.findAll('div', attrs = {"class":"lister-item mode-advanced"})
```

```
for store in movie_data:
    name = store.h3.a.text
    movie_name.append(name)
    
    movie_year = store.h3.find("span", class_ = "lister-item-year text-muted unbold").text.replace('(','').replace(')','')
    year.append(movie_year)
    
    time = store.p.find("span", class_ = "runtime").text.replace(" min","")
    runtime.append(time)
    
    rate = store.find('div', class_ = "inline-block ratings-imdb-rating").text.replace("\n","")
    rating.append(rate)
    
    mscore = store.find("span", class_ = "metascore").text.replace(" ","") if store.find("span", class_ = "metascore") else "NA"
    metascore.append(mscore)
    
    movie_genre = store.find('span','genre').findAll('a')
    genre.append(movie_genre)
    
    user_votes = store.find_all("span", attrs = {"name": "nv"})
    vote = user_votes[0].text
    votes.append(vote)
    
    income = user_votes[1].text if len(user_votes) > 1 else "NA"
    gross.append(income)
    
        
```

```
movie_Df = pd.DataFrame({'Movie Title': movie_name, 'Year of release': year,'Genre': genre, 'Runtime': runtime, 'Movie Rating': rating, 'Metascore': metascore, 'Votes': votes, 'Gross Income': gross})
```

```
import pandas as pd
import sqlite3

movie_Df = ({'Movie Title': movie_name, 
             'Year of release': year, 
             'Runtime': runtime, 
             'Movie Rating': rating, 
             'Metascore': metascore, 
             'Votes': votes, 
             'Gross Income': gross})


 

tableName   = "movie_Df18"

dataFrame   = pd.DataFrame(data=movie_Df)           

sqlEngine=sqlite3.connect('FYM1.db')



try:

    frame= dataFrame.to_sql(tableName, sqlEngine, if_exists='fail');

except ValueError as vx:

    print(vx)

except Exception as ex:   

    print(ex)

else:

    print("Table %s created successfully."%tableName);   

finally:

    sqlEngine.close()
```
